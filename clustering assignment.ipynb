```python


url = 'https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M'


df = pd.read_html(url)
```

selecting df of interest among the scraped page and filter results as instructed 

```python
df_1=df[0]
df_1=df_1[df_1['Borough']!='Not assigned']
df_1
```

appling the index reset and print the shape of the object

```python
df_1=df_1.reset_index()
df_1=df_1[['Postal Code', 'Borough', 'Neighbourhood']]
df_1.head()
```

```python
df_1.shape
```

```python
c_df=pd.read_csv('C:\\Users\\giuli\\JUPITER\\Geospatial_Coordinates.csv')
c_df
```

merging tables 

```python
p=pd.merge(df_1, c_df, how='inner', on=None, left_on=None, right_on=None,
         left_index=False, right_index=False, sort=True,
         suffixes=('_x', '_y'), copy=True, indicator=False,
         validate=None)
```

```python
p
```

display neighboors on the map 

```python
map_T = folium.Map(location=[43.6532, -79.3832], zoom_start=11)

# add markers to map
for lat, lng, borough, neighborhood in zip(p['Latitude'], p['Longitude'], p['Borough'], p['Neighbourhood']):
    label = '{}, {}'.format(neighborhood, borough)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        popup=label,
        color='blue',
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.7,
        parse_html=False).add_to(map_T)  
    
map_T
```
